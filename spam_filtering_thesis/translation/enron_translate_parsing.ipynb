```python
#Import the tools
# enron dataset
import docx
from google.colab import files

uploaded = files.upload()
file_name = 'enron_pre_translate3.docx'
doc = docx.Document(file_name)
# print(doc.paragraphs[1].text)






```

```python
import pandas as pd
txt = []
for i in range(1,len(doc.paragraphs)-1):
  txt.append(doc.paragraphs[i].text)



dataset = pd.DataFrame(txt, columns =['gtrans_el'])
print(dataset.loc[1])
dataset.to_csv('enron_translate3.csv', index=False)
files.download('enron_translate3.csv')


```
