```python
from google.colab import drive
from google.colab import files
import pandas as pd
from googletrans import Translator
drive.mount('/content/drive')  # prepei na anebasw to dataset sto gdrive
```

```python

data = pd.read_csv('/content/drive/MyDrive/spam_detection/sms_pre_translate.csv') # load spam dataset
flag = 'sms' # add the flag

# data = pd.read_csv('/content/drive/MyDrive/spam_detection/youtube_data.csv') # load youtube dataset
# flag = 'youtube' # add the flag
# print(data)

# data = pd.read_csv('/content/drive/MyDrive/spam_detection/custom_test_set.csv') # run this for custom spam dataset for evaluation
# flag = 'custom' # add the flag
# print(data)

```

```python
# after translation to remove this  '« " ' that occur after greek translation
import re

def clear(text):
  new_txt= re.sub(r'[\u00AB"\u00BB]', "", text ) # to remove this  '«' that occurs after greek translation
  return new_txt
```

```python
# Initialize the translator object
translator = Translator()

# function to translate the content of emails using google translate
def translate(text):
   result = translator.translate(text, src='en', dest='el')
   return result.text

# Translate
gtrans_el=[]

for i in data.index:
  gtrans_el.append(translate(data.loc[i,'Message']))
  print(i)

data['gtrans_el'] = gtrans_el
data.gtrans_el = data.gtrans_el.apply(clear)


if flag == 'sms':
    data.to_csv('sms_translate.csv', index=False)
    files.download('sms_translate.csv')
elif flag == 'youtube':
    data.to_csv('youtube_translate.csv', index=False)
    files.download('youtube_translate.csv')
elif flag == 'custom':
    data.to_csv('custom_test_set.csv', index=False)
    files.download('custom_test_set.csv')





```
